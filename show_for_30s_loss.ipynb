```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = r"../lib/attention_model_data/loss/loss_30s.csv"
loss_df = pd.read_csv(file)
```

```python
# 
%matplotlib inline

plt.figure(figsize=(12, 6))  # 

# 
plt.plot(loss_df.index, loss_df.iloc[:, 0], linewidth=1.5, color='steelblue')

# 
plt.title('MSELoss map', fontsize=16)
plt.xlabel('time', fontsize=12)
plt.ylabel('MSELoss', fontsize=12)

# 
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.tight_layout()

# 
plt.show()
```
